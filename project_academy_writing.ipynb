```python

import requests
from bs4 import BeautifulSoup
import nltk
from collections import Counter
import matplotlib.pyplot as plt


nltk.download("punkt")
nltk.download('averaged_perceptron_tagger')
```

```python

page = requests.get("https://applied-llms.org/")
paragraphs = BeautifulSoup(page.content).find_all("p")
page_text = " ".join(paragraph.getText() for paragraph in paragraphs)

counter = Counter()

for word, tag in nltk.pos_tag(nltk.word_tokenize(page_text)):
    counter[tag] += 1


print(counter)
```

```python
plt.rcParams['figure.figsize'] = [20, 5]
plt.bar(counter.keys(), counter.values())
```

```python
plt.rcParams['figure.figsize'] = [20, 20]
plt.pie(counter.values(), labels=counter.keys())
```
